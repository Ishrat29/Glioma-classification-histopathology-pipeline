# Mixed Stain Normalization

Blends Reinhard- and Macenko-normalized image sets via weighted linear combination (`alpha`). Includes variants for multiclass, necrosis, and binary datasets. Update paths before running.

```python
import numpy as np
import cv2
import os

# Define paths to your Reinhard and Macenko normalized datasets
reinhard_base_path = '<REINHARD_BASE_PATH>'
macenko_base_path = '<MACENKO_BASE_PATH>'
mixed_base_path = '<MIXED_OUTPUT_PATH>'

# Ensure the mixed dataset directory exists
if not os.path.exists(mixed_base_path):
    os.makedirs(mixed_base_path)

# Define the classes
classes = ['Astro', 'Gbm', 'Oligo']

# Define the mixing factor
alpha = 0.5  # You can adjust this value or randomize it between 0 and 1

# Process each class
for class_name in classes:
    # Create class directories in the mixed dataset path
    mixed_class_path = os.path.join(mixed_base_path, class_name)
    if not os.path.exists(mixed_class_path):
        os.makedirs(mixed_class_path)
    
    # Get paths to the current class in both datasets
    reinhard_class_path = os.path.join(reinhard_base_path, class_name)
    macenko_class_path = os.path.join(macenko_base_path, class_name)
    
    # List all patient folders in the current class directory
    patient_folders = os.listdir(reinhard_class_path)
    
    for patient_folder in patient_folders:
        # Create corresponding patient folder in the mixed dataset path
        mixed_patient_folder_path = os.path.join(mixed_class_path, patient_folder)
        if not os.path.exists(mixed_patient_folder_path):
            os.makedirs(mixed_patient_folder_path)
        
        # Get paths to the current patient folder in both datasets
        reinhard_patient_folder_path = os.path.join(reinhard_class_path, patient_folder)
        macenko_patient_folder_path = os.path.join(macenko_class_path, patient_folder)
        
        # List all image files in the current patient folder
        reinhard_files = os.listdir(reinhard_patient_folder_path)
        
        for file_name in reinhard_files:
            # Load the corresponding images
            reinhard_img_path = os.path.join(reinhard_patient_folder_path, file_name)
            macenko_img_path = os.path.join(macenko_patient_folder_path, file_name)
            
            reinhard_img = cv2.imread(reinhard_img_path)
            macenko_img = cv2.imread(macenko_img_path)

            # Ensure the images have the same shape
            if reinhard_img is None or macenko_img is None:
                print(f"Error loading images: {file_name} in {patient_folder}. Skipping.")
                continue
            
            assert reinhard_img.shape == macenko_img.shape, f"Image dimensions do not match for {file_name} in {patient_folder}!"

            # Perform the linear combination
            mixed_img = alpha * reinhard_img + (1 - alpha) * macenko_img

            # Clip the pixel values to the valid range (0 to 255 if using uint8 images)
            mixed_img = np.clip(mixed_img, 0, 255).astype(np.uint8)

            # Save the mixed image
            mixed_img_path = os.path.join(mixed_patient_folder_path, file_name)
            cv2.imwrite(mixed_img_path, mixed_img)

            print(f"Processed {file_name} in {patient_folder} for class {class_name}.")

print("Mixing of datasets completed.")
```
